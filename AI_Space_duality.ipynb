```python
!yolo task = detect mode = predict model = runs/detect/train2/weights/best.pt source = /content/drive/MyDrive/Hackathon_Dataset/HackByte_Dataset/data/test/images save = True
```

```python
from ultralytics import YOLO

# Load the trained model
model = YOLO('/content/drive/MyDrive/Hackathon_Dataset/HackByte_Dataset/runs/detect/train2/weights/best.pt')

# Validate the model on the test dataset and save the results
results = model.val(
    data='/content/drive/MyDrive/Hackathon_Dataset/HackByte_Dataset/yolo_params.yaml', # Assuming yolo_params.yaml contains the path to the test dataset
    split='test', # Specify to run validation on the test split
    save_json=True, # Save results to a json file
    save_hybrid=True, # Save hybrid version of labels
    plots=True, # Generate and save plots
    project='runs/detect',  # Specify the project directory
    name='val_on_test_data_new_run' # Specify a name for this new run
)
```
